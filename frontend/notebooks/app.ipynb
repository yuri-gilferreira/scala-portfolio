```python
import os
os.chdir('../../')

import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import plotly.express as px

```

```python
dataPath = Path('data')
df = pd.read_csv(dataPath/'simulations.csv', delimiter=';')
```

```python
df = df.sort_values(["Return"])
```

```python
fig = px.scatter(df, x='Risk', y='Return', text=df.index, hover_data=['Weights'])
# Update the layout if necessary
fig.update_layout(
    title='Risk vs. Return',
    xaxis_title='Risk',
    yaxis_title='Return',
    hovermode='closest'
)

# Show the figure
fig.show()
```
